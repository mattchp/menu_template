```python
import pandas as pd
import pymysql
from sqlalchemy import create_engine

# Enter your MySQL password to run this program
password = "xxx"

cnx = create_engine(f'mysql+pymysql://root:{password}@127.0.0.1/sys')
```

```python
import time
from datetime import date
today = date.today()
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
figure(figsize=(8, 6), dpi=80)
import numpy as np

def menu():
    print("--------------------")
    print("Trade System Tools")
    print("--------------------")
    print("[1] Query")
    print("[2] Export")
    print("[3] Report")
    print("[4] Exit")
    
menu()
choice = int(input("Please choose an option <1>, <2>, <3>, <4>: "))

while choice != 4:
    if choice == 1:
        print("Choice 1 has been called. Please follow the instruction below.")
        def sub_menu1():
            print("--------------------")
            print("Option 1: Query")
            print("--------------------")
            print("[1] List all Brokers")
            print("[2] List all Shares")
            print("[3] Lookup Trades by Trade ID")
            print("[4] Return to Main Menu")
        
        sub_menu1()
        sub_choice1 = int(input("Please input an option <1>, <2>, <3>, <4>: "))
        
        while sub_choice1 != 4:
            if sub_choice1 == 1:
                print("Sub Choice 1 has been called. Please follow the instruction below.")
                list_brokers = pd.read_sql('select * from brokers', cnx)
                print(list_brokers)
                
            elif sub_choice1 == 2:
                print("Sub Choice 2 has been called. Please follow the instruction below.")
                list_shares = pd.read_sql('''select c.name, s.share_id from companies c 
                inner join shares s on c.company_id = s.company_id order by s.share_id''', cnx)
                print(list_shares)
                
            elif sub_choice1 == 3:
                print("Sub Choice 3 has been called. Please follow the instruction below.")
                list_trades = pd.read_sql('select * from trades order by trade_id', cnx)
                print(list_trades.to_string())
            
            elif sub_choice1 == 4:
                break
            
            else:
                print("Invalid Option: Please Try Again")
            
            print()    
            sub_menu1()
            sub_choice1 = int(input("Please input an option <1>, <2>, <3>, <4>: "))
        
    elif choice == 2:
        print("Choice 2 has been called. Please follow the instruction below.")
        def sub_menu2():
            print("--------------------")
            print("Option 2: Export")
            print("--------------------")
            print("[1] Fetch Trades by Broker ID")
            print("[2] Fetch Trades by Trade ID")
            print("[3] Return to Main Menu")
        
        sub_menu2()
        sub_choice2 = int(input("Please input an option <1>, <2>, <3>: "))
        
        while sub_choice2 != 3:
            if sub_choice2 == 1:
                print("Sub Choice 1 has been called. Please follow the instruction below.")
                trades_broker_id = pd.read_sql('select * from trades order by broker_id', cnx)
                print(trades_broker_id)
                
                time.sleep(2)
                print("Fetching data to csv file...")
                trades_broker_id.to_csv('tradesbybrokerid.csv',index=False)
                time.sleep(2)
                print("The data has been fetched to your working directory.")
                
            elif sub_choice2 == 2:
                print("Sub Choice 2 has been called. Please follow the instruction below.")
                trades_broker_id = pd.read_sql('select * from trades order by trade_id', cnx)
                print(trades_broker_id)
                
                time.sleep(2)
                print("Fetching data to csv file...")
                trades_broker_id.to_csv('tradesbytradeid.csv',index=False)
                time.sleep(2)
                print("The data has been fetched to your working directory.")
                
            elif sub_choice2 == 3:
                break
            
            else:
                print("Invalid Option: Please Try Again")
            
            print()    
            sub_menu2()
            sub_choice2 = int(input("Please input an option <1>, <2>, <3>: "))
        
    elif choice == 3:
        print("Choice 3 has been called. Please follow the instruction below.")
        def sub_menu3():
            print("--------------------")
            print("Option 3: Report")
            print("--------------------")
            print("[1] Number of Trades Per Broker [Bar Chart]")
            print("[2] Proportion of Trades Traded on Each Exchange [Pie Chart]")
            print("[3] Return to Main Menu")
            
        sub_menu3()
        sub_choice3 = int(input("Please input an option <1>, <2>, <3>: "))
        
        while sub_choice3 != 3:
            if sub_choice3 == 1:
                print("Sub Choice 1 has been called. Please follow the instruction below.")
                num_trade_broker = pd.read_sql('''select concat(b.first_name, ' ', b.last_name) as Broker, 
                count(t.trade_id) as 'Number of Trades' from brokers b inner join trades t on b.broker_id = 
                t.broker_id group by concat(b.first_name, ' ', b.last_name)''', cnx)
                
                broker = num_trade_broker['Broker']
                num_trade = num_trade_broker['Number of Trades']
                
                figure(figsize=(8, 6), dpi=80)
                plt.xlabel('Broker')
                plt.ylabel('Number of Trades')
                plt.title('Number of Trades for every Broker')
                plt.bar(broker, num_trade)
                plt.show()
                
                print(num_trade_broker)
                
            elif sub_choice3 == 2:
                print("Sub Choice 2 has been called. Please follow the instruction below.")
                traded_exchange = pd.read_sql('''select se.name as 'Stock Exchange Name', count(t.trade_id) 
                as 'Number of Trades' from stock_exchanges se inner join trades t on se.stock_ex_id = 
                t.stock_ex_id group by se.name''', cnx)
                
                se_name = traded_exchange['Stock Exchange Name'].to_numpy()
                print(se_name)
                se_lon = se_name[0]
                se_par = se_name[1]
                se_nyc = se_name[2]
                
                num_trade2 = traded_exchange['Number of Trades'].to_numpy()
                print(num_trade2)
                trade_lon = num_trade2[0]
                trade_par = num_trade2[1]
                trade_nyc = num_trade2[2]
                
                tot_lon = np.array([num_trade2[0]])
                tot_par = np.array([num_trade2[1]])
                tot_nyc = np.array([num_trade2[2]])
                
                figure(figsize=(8, 6), dpi=80)
                lon_full = np.sum(tot_lon)
                par_full = np.sum(tot_par)
                nyc_full = np.sum(tot_nyc)
                total_trades = np.array([lon_full, par_full, nyc_full])
                pie_labels = [se_lon, se_par, se_nyc]
                plt.pie(total_trades, labels=pie_labels, autopct='%.2f%%', shadow=True)
                plt.title('Proportion of Trades Traded on Each Exchange')
                plt.show()                
                
                print(traded_exchange)
                
                #combined = np.vstack((se_name, num_trade2)).T
                #print(combined)                
                #total_trade = sum(num_trade2)
                #print(total_trade)
                
                
            elif sub_choice3 == 3:
                break
            
            else:
                print("Invalid Option: Please Try Again")
            
            print()    
            sub_menu3()
            sub_choice3 = int(input("Please input an option <1>, <2>, <3>: "))
        
    else:
        print("Invalid Option: Please Try Again")
    
    print()    
    menu()
    choice = int(input("Please choose an option <1>, <2>, <3>, <4>: "))
    
print("Thanks for using the trading system. Goodbye.")
```
